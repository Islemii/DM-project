```python
# v√©rifier s'il existe des valeurs manquantes dans le dataset
df.isnull().sum()
```

Il ya 3829 lignes qui conteint comme genre NaN

```python
# Remplacement des valeurs NaN dans la colonne "gender"
# D√©finition d'une fonction pour d√©terminer le genre en fonction du nom
def determine_gender(row):
    if pd.isna(row['gender']):
        if row['full_name'] == "Bill Gates":
            return "Male"  # Si le nom complet est "Bill Gates", on met "Male"
    else:
        return row['gender']  # Si le genre est d√©j√† d√©fini, on le conserve

# Application de la fonction determine_gender pour remplacer les valeurs NaN dans la colonne "gender"
df['gender'] = df.apply(determine_gender, axis=1)
```

```python
filtered_rows = df[df['full_name'] == "Bill Gates"]
print(filtered_rows)
```

```python
#Suppression des lignes qui contiennent nulle dans la colonne Genre
df=df.dropna(subset=['gender'])
df.shape
```

```python
#enl√©ver B de net_woth et rendre le type float
df['net_worth'] = df['net_worth'].str.replace('B',"").astype(float) 
```

```python
df.head()
```

```python
# Liste des colonnes √† supprimer
colonnes_inutiles = ['month','last_name','organization_name','country_of_citizenship','city_of_residence','business_industries']

# Supprimer les colonnes inutiles pour mon analyse
df = df.drop(columns=colonnes_inutiles)
```

```python
# Pour les valeurs manquantes dans la colonne "first_name", il les remplace par le premier mot de la colonne "full_name"
df.loc[list(df.first_name.isnull().index),'first_name'] = df.loc[list(df.first_name.isnull().index),'full_name'].str.split().apply(lambda x:x[0])
```

```python
# Convert birth_date to datetime (replace with your date format)
df['birth_date'] = pd.to_datetime(df['birth_date'], format='%Y-%m-%d')
```

```python
df.info()
```

```python
# Calculer la moyenne des √¢ges
mean_age = df['age'].mean()

# Remplacer les valeurs nulles de la colonne "age" par la moyenne
df['age'].fillna(mean_age, inplace=True)
```

```python
# Supprimer les lignes contenant des valeurs nulles dans la colonne "rank"
df.dropna(subset=['rank'], inplace=True)
df.info()
```

```python
# Select duplicate rows except first occurrence based on all columns
duplicateRowsDF = df[df.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
duplicateRowsDF.shape
```

```python
#Supprimer les lignes Duppliqu√©es
df=df.drop_duplicates()
df.shape
```

```python
df.info()
```

```python

# Cr√©er le dictionnaire pour mapper les valeurs de genre
M_vs_F = {'Male': 0, 'Female': 1 , '1':0, '2':1 }

# Afficher les valeurs uniques dans la colonne 'gender'
print(df['gender'].unique())

# Remplacer les valeurs non d√©finies dans le dictionnaire par une valeur par d√©faut (par exemple, -1)
df['gender'].fillna(-1, inplace=True)

# Utiliser le dictionnaire pour mapper la colonne 'gender' √† la nouvelle colonne 'gender_0_1'
df['gender'] = df['gender'].map(M_vs_F)

# Convertir la colonne en type 'int64'
df['gender'] = df['gender'].astype('int64')

# Afficher les premi√®res lignes pour v√©rification
df.head()
```

## <font color='red'>Interpr√©tation:
Mean Squared Error (MSE) : Cette m√©trique mesure en moyenne la diff√©rence quadratique entre les valeurs pr√©dites par le mod√®le et les valeurs r√©elles. Dans votre cas, le MSE est de 0.0347. Cela signifie que, en moyenne, les pr√©dictions de votre mod√®le diff√®rent des valeurs r√©elles de 0.0347 unit√©s au carr√©. √âtant donn√© que le MSE est une mesure quadratique, il est difficile de l'interpr√©ter directement en unit√©s de la variable cible, mais une valeur plus basse indique g√©n√©ralement de meilleures performances.
Mean Absolute Error (MAE) : Cette m√©trique mesure en moyenne la diff√©rence absolue entre les valeurs pr√©dites par le mod√®le et les valeurs r√©elles. Dans votre cas, le MAE est de 0.0045. Cela signifie que, en moyenne, les pr√©dictions de votre mod√®le diff√®rent des valeurs r√©elles de 0.0045 unit√©s. Comme pour le MSE, une valeur plus basse indique de meilleures performances.
R-squared (Coefficient de d√©termination) : Ce coefficient mesure la proportion de la variance de la variable d√©pendante (net_worth) qui est expliqu√©e par le mod√®le.
